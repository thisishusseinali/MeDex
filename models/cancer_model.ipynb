```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import pickle
import joblib
```

```python
cancer_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Graduation Project/TheClinic/datasets /cancer.csv')
cancer_data.head()
```

```python
cancer_data.drop(["Unnamed: 32"],axis="columns",inplace=True)
```

```python
cancer_data.drop(["id"],axis="columns",inplace=True)
```

```python
diagnosis_dummies = pd.get_dummies(cancer_data["diagnosis"])
```

```python
cancer_data = pd.concat([cancer_data,diagnosis_dummies],axis="columns")
cancer_data.head()
```

```python
cancer_data.drop(["diagnosis","B"],axis="columns",inplace=True)
cancer_data.rename(columns={"M":"Malignant/Benign"},inplace=True)
cancer_data.head()
```

```python
Y=cancer_data[["Malignant/Benign"]]
X=cancer_data.drop(["Malignant/Benign"],axis="columns")
print(X.shape[1])
print(Y.shape[1])
```

```python
x=np.array(X)
y=np.array(Y)
```

```python
cancer_model = LogisticRegression()
cancer_model.fit(x,y.reshape(-1,))
```

```python
joblib.dump(cancer_model,"cancer_model")
```
